## Handling Multiple sequences

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python

sequence = "This restaurant has the best food in town"

tokens = tokenizer.tokenize(sequence)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(tokens)
ids
```

```python
input_ids = torch.tensor([ids])
input_ids
```

```python
output = model(input_ids)
# print(output)
output.logits
```

```python
torch.nn.functional.softmax(output.logits, dim=-1)
```

**When you’re trying to batch together two (or more) sentences, they might be of different lengths. If you’ve ever worked with tensors before, you know that they need to be of rectangular shape, so you won’t be able to convert the list of input IDs into a tensor directly**

```python
batched_ids = [
    [200, 200, 200],    #  ValueError: expected sequence of length 3 at dim 1 (got 2)
    [200, 200]
]

# we’ll use padding to make our tensors have a rectangular shape.
# Padding makes sure all our sentences have the same length by adding a special word called the padding token to the sentences with fewer values
batched_ids = [
    [200, 200, 200],
    [200, 200,0]
]
print(model(torch.tensor(batched_ids)).logits)
batched_ids = [
    [200, 200, 200],
    [200, 200,100]
]
model(torch.tensor(batched_ids)).logits
```

 To get the same result when passing individual sentences of different lengths through the model or when passing a batch with the same sentences and padding applied, we need to tell those attention layers to ignore the padding tokens. This is done by using an attention mask.

```python
batched_ids = [
    [200, 200, 200],
    [200, 200,0]
]
attention_mask = [
    [1, 1, 1],
    [1, 1, 0],
]
print(model(torch.tensor(batched_ids),torch.tensor(attention_mask)).logits)
batched_ids = [
    [200, 200, 200],
    [200, 200,200]
]
attention_mask = [
    [1, 1, 1],
    [1, 1, 0],
]
model(torch.tensor(batched_ids),torch.tensor(attention_mask)).logits
```

Now we get the same logits

**Attention masks are tensors with the exact same shape as the input IDs tensor, filled with 0s and 1s: 1s indicate the corresponding tokens should be attended to, and 0s indicate the corresponding tokens should not be attended to (i.e., they should be ignored by the attention layers of the model)**

## Putting it all together

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

model_inputs = tokenizer(sequence)
model_inputs
```

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

model_inputs = tokenizer(sequences)
model_inputs
```

```python
# Will pad the sequences up to the maximum sequence length
model_inputs = tokenizer(sequences, padding="longest")
print(len(model_inputs.input_ids[1]))
print(model_inputs)
# Will pad the sequences up to the model max length
# (512 for BERT or DistilBERT)
model_inputs = tokenizer(sequences, padding="max_length")
print(len(model_inputs.input_ids[1]))
print(model_inputs)

# Will pad the sequences up to the specified max length
model_inputs = tokenizer(sequences, padding="max_length", max_length=8)
print(len(model_inputs.input_ids[1]))
model_inputs
```

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

# Will truncate the sequences that are longer than the model max length
# (512 for BERT or DistilBERT)
# If the sequence length is less than or equal to the max length, truncation will not occur.
model_inputs = tokenizer(sequences, truncation=True)
print(len(model_inputs.input_ids[0]))
print(model_inputs)

# Will truncate the sequences that are longer than the specified max length
model_inputs = tokenizer(sequences, max_length=8, truncation=True)
print(len(model_inputs.input_ids[0]))
model_inputs
```

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

# Returns PyTorch tensors
model_inputs = tokenizer(sequences, padding=True, return_tensors="pt")
print('PyTorch tensors')
print(model_inputs)
print("-"*100)

# Returns TensorFlow tensors
model_inputs = tokenizer(sequences, padding=True, return_tensors="tf")
print('TensorFlow tensors')
print(model_inputs)
print("-"*100)

# Returns NumPy arrays
model_inputs = tokenizer(sequences, padding=True, return_tensors="np")
print('NumPy tensors')
model_inputs
```
